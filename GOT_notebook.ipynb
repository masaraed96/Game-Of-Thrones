## Loading Datasets

```python
#read JSON file
#convert JSON file to Pandas DataFrame
def json_to_dataframe(JSON_file, stringname):
    with open(JSON_file, "r") as file:
        file_load = json.load(file)
    return pd.json_normalize(file_load[stringname])
```

## Analyzing and Visualizing Data

### Q1 : Who's the most acted character in all seasons?

```python
characters['number_of_occurenece'] = 0
d = dict() 
for characters_names in scenes.characters_names:
    for name in characters_names: 
        if name in d: 
            d[name] = d[name] + 1
        else: 
            d[name] = 1
```

```python
for name in d:
    characters.loc[characters['characterName'] == name, 'number_of_occurenece'] = d[name]
```

```python
characters = characters.sort_values(by='number_of_occurenece', ascending=False)
characters[:14].plot(kind = 'barh',x='characterName', y='number_of_occurenece')
plt.title('Most acted characters')
plt.xlabel('Count')
plt.ylabel('character name');
```

### Q4: type of scenes in episodes?

```python
scenes['scene_type'] = ""
#create a column for names of characters in each scene
for index, scene in scenes.iterrows():
    _types = scene.characters
    types = []
    for key in _types:
        for x in key.keys():
            if((x not in ['name', 'title']) & (x not in types)):
                types.append(x)
    if(len(types) == 0):
        scenes.at[index, 'scene_type'] = np.nan
    else:
        scenes.at[index, 'scene_type'] = str(types)
```

```python
scenes.scene_type = scenes.scene_type.astype(str)
```
